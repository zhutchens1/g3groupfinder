```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import mode
```

```python
g3 = pd.read_csv("../resolve_and_eco/ECOdata_G3catalog_luminosity.csv")
```

```python
g3coma=g3[g3.g3grp_l==13]
```

```python
fof = pd.read_csv("ECODR3_Jul0822.csv")
```

```python
fofcoma = fof[fof.grp==26]
```

```python
len(fofcoma), len(g3coma)
```

```python
plt.figure()
plt.plot(g3coma.radeg,g3coma.dedeg, 's', markerfacecolor='None', markeredgecolor='k', label='G3')
plt.plot(fofcoma.radeg,fofcoma.dedeg, 'r.', label='FoF', markersize=3)
plt.xlabel("RA [deg]")
plt.ylabel("Dec [deg]")
plt.legend(loc='best')
plt.show()
```

```python
plt.figure()
plt.plot(g3coma.radeg,g3coma.cz, 's', markerfacecolor='None', markeredgecolor='k', label='G3')
plt.plot(fofcoma.radeg,fofcoma.cz, 'r.', label='FoF', markersize=3)
plt.xlabel("RA [deg]")
plt.ylabel("cz [km/s]")
plt.legend(loc='best')
plt.show()
```

```python
plt.figure()
plt.plot(g3coma.dedeg,g3coma.cz, 's', markerfacecolor='None', markeredgecolor='k', label='G3')
plt.plot(fofcoma.dedeg,fofcoma.cz, 'r.', label='FoF', markersize=3)
plt.xlabel("Dec [deg]")
plt.ylabel("cz [km/s]")
plt.legend(loc='best')
plt.show()
```
