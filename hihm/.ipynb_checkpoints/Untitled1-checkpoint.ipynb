```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
eco = pd.read_csv("../resolve_and_eco/ECOdata_G3catalog_luminosity.csv")
```

```python
eco = eco[(eco.mhidet_a100>0)&(eco.mhidet>0)&(eco.resname!='notinresolve')]
```

```python
fig,axs=plt.subplots(ncols=2)
perr = (eco.mhidet_a100-eco.mhidet)/(eco.mhidet)*100
axs[0].plot(eco.mhidet, perr,'k.')
axs[0].set_xlabel("HI Mass from RESOLVE")
axs[0].set_ylabel("Percent Error between RESOLVE and\nALFALFA-100 HI Masses")
axs[0].set_ylim(-100,100)
axs[0].set_xscale('log')
axs[0].set_xlim(5e8,5e10)
axs[1].hist(perr,np.arange(-100,100,2),orientation='horizontal',log=True)
axs[1].set_xlabel("# of galaxies", fontsize=12)
plt.show()
```

```python
plt.hist(np.abs(perr),bins=np.arange(0,100,1))
plt.yscale('log')
```
