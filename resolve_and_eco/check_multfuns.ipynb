```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
import numpy as np
sys.path.insert(0,'../g3algo/')
from foftools import multiplicity_function
```

```python
ecodr2=pd.read_csv("ECODR2.csv")
resolve = pd.read_csv("RESOLVEdata_080822.csv")
```

```python
ecodr2=ecodr2[ecodr2.absrmag<-17.33]
```

```python
resolve = resolve[((resolve.f_b>0)&(resolve.absrmag<-17))]
```

```python
resolve=resolve[(resolve.grp_e17>0)&(resolve.f_b>0)]
```

```python
plt.figure()
binv=np.arange(0.5,300.5,1)
plt.hist(multiplicity_function(ecodr2.grp_e17,return_by_galaxy=False), log=True,bins=binv)
plt.hist(multiplicity_function(resolve.grp_e17,return_by_galaxy=False), log=True,bins=binv)
plt.ylim(0.5,1e4)
plt.xlim(0,30)
plt.show()
```
